### One Hot Encoding
* Encodage de la variable city par la methode du one hot encoding 

```python
## encodage de la variable  qualitative city

var_Qual_Encode = pd.get_dummies(var_Qual)
vecteur = range(len(var_num))
var_Qual_Encode = var_Qual_Encode.assign(nouvelle_variable = vecteur)
print(f'var_Qual_Encode.shape: {var_Qual_Encode.shape}')
var_Qual_Encode.head()
```

### Standardizing the Data

* Standardisation de X par la methode du RobusteScaler

```python
## standardization des variables quantitatives (var_num)

scaler = RobustScaler()
var_num_standar = scaler.fit_transform(var_num)
var_num_standar = pd.DataFrame(var_num_standar, columns= ['bedrooms', 'bathrooms', 'sqft_living', 'floors', 'waterfront', 'view',
       'sqft_above', 'sqft_basement'])
var_num_standar = var_num_standar.assign(nouvelle_variable = vecteur)
print(f'var_num_standar.shape:{var_num_standar.shape}')
var_num_standar.head()
```

```python
## jointure de var_num_standar et var_Qual_Encode pour creer la base final pour la mod√©lisation 

X = pd.merge(var_num_standar, var_Qual_Encode, on='nouvelle_variable')
print(f'X.shape:{X.shape}')
X.head()
```

```python
X = X.drop('nouvelle_variable', axis = 1)
print(f'X.shape:{X.shape}')
X.head()
```

```python
## standardization de log_y

Y = scaler.fit_transform(log_y)
pd.DataFrame(Y).T
```

```python
## identification des train_set et test_set

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state = 56 )
print(f'X_train.shape: {X_train.shape}\nX_test.shape:{X_test.shape}\nY_train.shape:{Y_train.shape}\nY_test.shape:{Y_test.shape}')
```

## III-MODELISATION
* Nous utiliserons les quelsques mod√®les du Linear_Model(LinearRegression, Lasso, Ridge, ElasticNet) et quelsques mod√®les ensemblistes(RandomForest, XGBoost). Une comparaison sera faite pour ensuite choisir le meilleur mod√®le.

```python
def rmse_cv(model):
    rmse = np.sqrt(-cross_val_score(model, X_train, Y_train, scoring="neg_mean_squared_error", cv=5)).mean()
    return rmse
    

def evaluation(y, predictions):
    mae = mean_absolute_error(y, predictions)
    mse = mean_squared_error(y, predictions)
    rmse = np.sqrt(mean_squared_error(y, predictions))
    r_squared = r2_score(y, predictions)
    return mae, mse, rmse, r_squared
```

```python
## le modele

models=pd.DataFrame(columns=["Model","MAE","MSE","RMSE","R2 Score","RMSE (Cross-Validation)"])
```

## Linear_Model Methods

### Linear Model Regression

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)
lin_reg_pred = lin_reg.predict(X_test)

mae, mse, rmse, r_squared = evaluation(Y_test, lin_reg_pred)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r_squared)
print("-"*30)
rmse_cross_val = rmse_cv(lin_reg)
print("RMSE Cross-Validation:", rmse_cross_val)

new_row = {"Model": "LinearRegression","MAE": mae, "MSE": mse, "RMSE": rmse, "R2 Score": r_squared, "RMSE (Cross-Validation)": rmse_cross_val}
models = models.append(new_row, ignore_index=True)
```
